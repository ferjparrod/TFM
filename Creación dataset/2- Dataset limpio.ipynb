```python
# Cargamos algunas librerías necesarias.
import numpy  as np  
import pandas as pd
pd.set_option('display.max_columns', 100) # Con esto garantizo el ver todas las columnas.

from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt # Para dibujar alguna gráfica.
%matplotlib inline
from pygeocoder import Geocoder
from geopy.geocoders import Nominatim
import geocoder
import requests
import re
import threading 
import time
```

```python
df = pd.read_csv('./dfsemifinal_Accidentalidad.csv',sep=';', decimal='.', encoding = 'utf-8')
df.head()
```

```python
df
```

```python
# Borramos columnas que no vamos a necesitar.
df.drop(['Unnamed: 0', 'CruceCalle', 'Cruce', 'Calle completa', 'Cruces', 'point2', 'Autovia', 'point3'], axis=1, inplace=True)
```

```python
df.shape
```

```python
# Quitamos los Nan
df = df.dropna()
```

```python
df = df.reset_index(drop=True)
```

```python
df.shape
```

```python
df
```

```python
# Partimos el df en 2 columnas con la coma.
new = df["point"].str.split(",", n = 1, expand = True) 
  
# La primera columna será la latitud 
df["Latitude"]= new[0] 
  
# Y la segunda es la longitud 
df["Longitude"]= new[1] 
  
# Borramos nuestra columna
df.drop(columns =["point"], inplace = True) 
  
# Mostramos el dataset 
df
```

Vemos que hay que limpiarlo un poco más

```python
df['Latitude'] = df['Latitude'].str.extract(r'(\d+.\d+)').astype('float')
df['Latitude']
```

```python
df['Longitude'] = df['Longitude'].str.extract(r'(\d+.\d+)').astype('float')
df['Longitude'] = df['Longitude'] * -1 # Para volver a hacerlo negativo
df['Longitude']
```

```python
df
```

```python
# Los redondeo a 7 decimales porque había algunos campos que excedían esa cifra.
df['Latitude'] = df['Latitude'].round(7)
df['Longitude'] = df['Longitude'].round(7)
```

```python
# Lo guardamos.
df.to_csv('./Accidentes.csv',sep=';', decimal='.')
```
